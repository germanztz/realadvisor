```python
import pandas as pd


informe = pd.read_csv('../datasets/gen_informe.csv')
```

```python
from report_generator import ReportGenerator
from realty_report import RealtyReport

# Crear una lista de objetos RealtyReport por cada fila del dataframe
property_reports = []
for _, row in informe.iterrows():
    report = RealtyReport(
        link=row['link'],
        type_v=row['type_v'],
        address=row['address'],
        town=row['town'], 
        price=row['price'],
        price_old=row['price_old'],
        info=row['info'],
        description=row['description'],
        tags=row['tags'],
        agent=row['agent'],
        created=row['created'],
        n_hab=row['n_hab'],
        sup_m2=row['sup_m2'],
        disponibilidad=row['disponibilidad'],
        precio_venta_stars=row['precio_venta_stars'],
        precio_alquiler_estimado=row['precio_alquiler_estimado'],
        precio_venta_estimado=row['precio_venta_estimado'],
        barrio=row['barrio'],
        sup_nombre=row['sup_nombre'],
        tipo=row['tipo'],
        precio_venta_10y=row['precio_venta_10y'],
        precio_venta_1y=row['precio_venta_1y'],
        precio_venta_5y=row['precio_venta_5y'],
        precio_alquiler_10y=row['precio_alquiler_10y'],
        precio_alquiler_1y=row['precio_alquiler_1y'],
        precio_alquiler_5y=row['precio_alquiler_5y'],
        rentabilidad_10y=row['rentabilidad_10y'],
        rentabilidad_1y=row['rentabilidad_1y'],
        rentabilidad_5y=row['rentabilidad_5y'],
        grow_acu_alquiler_10y=row['grow_acu_alquiler_10y'],
        grow_acu_venta_10y=row['grow_acu_venta_10y'],
        grow_acu_alquiler_1y=row['grow_acu_alquiler_1y'],
        grow_acu_venta_1y=row['grow_acu_venta_1y'],
        grow_acu_alquiler_5y=row['grow_acu_alquiler_5y'],
        grow_acu_venta_5y=row['grow_acu_venta_5y'],
        precio_m2=row['precio_m2'],
        precio_desv_media=row['precio_desv_media']
    )
    property_reports.append(report)

# Crear el generador de informes
generator = ReportGenerator()

# Generar el informe
html_path, pdf_path = generator.generate_report(property_reports[1])
print(f"Informe generado en:\nHTML: {html_path}\nPDF: {pdf_path}")
```
